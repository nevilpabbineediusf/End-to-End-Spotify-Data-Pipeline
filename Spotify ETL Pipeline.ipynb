```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
```

```python
client_credentials_manager = SpotifyClientCredentials(client_id = " ", client_secret = "")
print(client_credentials_manager.get_access_token(as_dict=False))
```

```python
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
playlist_link = "https://open.spotify.com/playlist/6VOedaf3eNWDOVpa9Qdlvg"
```

```python
playlist_URI = playlist_link.split("/")[-1]
playlist_URI
```

```python
data = sp.playlist_tracks(playlist_URI)
```

```python
album_list = []
for row in data['items']:
    album_id = row['track']['album']['id']
    album_name = row['track']['album']['name']
    release_date = row['track']['album']['release_date']
    total_tracks = row['track']['album']['total_tracks']
    url = row['track']['album']['external_urls']['spotify']
    element = {'album_id' : album_id, 'album_name' : album_name, 'release_date' : release_date, 'total_tracks' : total_tracks, 'url' : url}
    album_list.append(element)
```

```python
data['items'][0]
```

```python
artist_list = []
for row in data['items']:
    for track in row['track']['artists']:
       artist_dict = {'artist_id' : track['id'], 'artist_name' : track['name'], 'external_url' : track['external_urls']['spotify']}
       artist_list.append(artist_dict)
```

```python
artist_list
```

```python
data['items'][0]['track']
```

```python
song_list = []
for row in data['items']:
    song_id = row['track']['id']
    song_name = row['track']['name']
    song_duration = row['track']['duration_ms']
    song_url = row['track']['external_urls']['spotify']
    song_popularity = row['track']['popularity']
    song_added = row['added_at']
    album_id = row['track']['album']['id']
    artist_id = row['track']['album']['artists'][0]['id']
    song_element = {'song_id':song_id,'song_name':song_name,'duration_ms':song_duration,'url':song_url,
                    'popularity':song_popularity,'song_added':song_added,'album_id':album_id,
                    'artist_id':artist_id
                   }
    song_list.append(song_element)
```

```python
song_list
```

```python
album_df = pd.DataFrame.from_dict(album_list)
```

```python
artist_df = pd.DataFrame.from_dict(artist_list)
```

```python
song_df = pd.DataFrame.from_dict(song_list)
```
